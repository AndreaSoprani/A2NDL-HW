```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

cwd = os.getcwd()
```

```python
dataset_dir = os.path.join(cwd, "Classification_Dataset")
img_files = []

for (root,dirs,files) in os.walk(dataset_dir):
    for file in files:
        img_files.append(os.path.join(root, file))
   
dims = []
for file in img_files:
    img = Image.open(file).convert("RGB")
    dims.append(img.size)
    
len(dims)
```

```python
dims = np.array(dims)
print("Image with max width: " + str(dims[np.argmax(dims[:,0])]))
print("Image with max height: " + str(dims[np.argmax(dims[:,1])]))
print("Image with min width: " + str(dims[np.argmin(dims[:,0])]))
print("Image with min height: " + str(dims[np.argmin(dims[:,1])]))
```

```python
plt.figure(figsize=(20, 10))
plt.plot(dims[:,0], dims[:,1], "ro", markersize = 3)
plt.grid()
plt.show()
```
